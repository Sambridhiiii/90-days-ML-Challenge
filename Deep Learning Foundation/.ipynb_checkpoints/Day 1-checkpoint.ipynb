# 1Ô∏è‚É£ Deep Learning Foundations

## üîπ What is Feature Engineering?
Feature Engineering is the process of **transforming raw data** into **meaningful features** that improve the performance of Machine Learning models.

- A **feature** (attribute) = measurable property of data.
- Example: For predicting house prices, features could be `size`, `location`, `age of house`.

---

## üîπ Why is Feature Engineering Important?
- ‚úÖ **Improves accuracy** ‚Üí Models learn better with good features.
- ‚úÖ **Reduces overfitting** ‚Üí Focus on relevant features only.
- ‚úÖ **Enhances interpretability** ‚Üí Models become easier to understand.
- ‚úÖ **Speeds up training** ‚Üí Clean, well-scaled data helps algorithms run faster.

---

## üîπ Where is Feature Engineering Used?
Feature engineering is applied **before training ML models** in various domains:
- üè† Real Estate: price prediction (create `price per sqft`, `house age`).
- üì∑ Computer Vision: edge detection, color histograms.
- üìù NLP (Text Data): TF-IDF, sentiment scores, word embeddings.
- üè• Healthcare: BMI from `weight/height¬≤`, risk factor scores.
- üìà Finance: ratios like `debt-to-income`, moving averages.

---

## üîπ When Do We Use Feature Engineering?
- When **raw data** is not suitable for direct input to ML models.
- Before applying algorithms like **SVM, Random Forest, Logistic Regression** (which need well-prepared features).
- Even in **deep learning**, engineered features (e.g., normalization, embeddings) help improve results.

---

## üîπ Key Processes
1. **Feature Creation** ‚Üí Build new features (`price_per_age`).
2. **Feature Transformation** ‚Üí Normalize, log-transform skewed data.
3. **Feature Extraction** ‚Üí PCA, dimensionality reduction.
4. **Feature Selection** ‚Üí Choose most relevant features.
5. **Feature Scaling** ‚Üí Standardization, Min‚ÄìMax scaling.

---

## üîπ Common Techniques
- **One-Hot Encoding** ‚Üí Convert categorical variables to numeric.
- **Binning** ‚Üí Group continuous values into ranges.
- **Text Preprocessing** ‚Üí Stopword removal, stemming, vectorization.
- **Feature Splitting** ‚Üí Extract info from complex features (e.g., splitting dates into day/month/year).

---

## üîπ Tools
- ‚öôÔ∏è **Featuretools** ‚Üí Automated feature engineering.
- ‚öôÔ∏è **TPOT** ‚Üí Genetic programming for feature selection.
- ‚öôÔ∏è **DataRobot / H2O.ai** ‚Üí AutoML with feature engineering.
- ‚öôÔ∏è **Alteryx** ‚Üí Drag-and-drop feature cleaning/extraction.

---

## ‚úÖ Summary
- **Why?** ‚Üí To improve model accuracy, interpretability, and efficiency.  
- **Where?** ‚Üí In all ML domains (text, image, finance, healthcare).  
- **When?** ‚Üí Before training models, especially traditional ML models.  


```python
# ==========================
# Import Libraries
# ==========================
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA

# ==========================
# Sample Dataset
# ==========================
data = {
    'Age': [25, 32, 47, 51, 62],
    'Salary': [40000, 52000, 80000, 110000, 150000],
    'City': ['Kathmandu', 'Pokhara', 'Kathmandu', 'Lalitpur', 'Pokhara']
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# ==========================
# 1. Feature Creation
# ==========================
# Create a new feature: Salary per Age
df['Salary_per_Age'] = df['Salary'] / df['Age']

# ==========================
# 2. Feature Transformation
# ==========================
# Log transformation to reduce skewness
df['Log_Salary'] = np.log(df['Salary'])

# ==========================
# 3. Feature Encoding (Categorical ‚Üí Numeric)
# ==========================
encoder = OneHotEncoder(sparse_output=False)
city_encoded = encoder.fit_transform(df[['City']])
city_df = pd.DataFrame(city_encoded, columns=encoder.get_feature_names_out(['City']))
df = pd.concat([df, city_df], axis=1)

# ==========================
# 4. Feature Scaling
# ==========================
scaler = StandardScaler()
df[['Age_scaled', 'Salary_scaled']] = scaler.fit_transform(df[['Age', 'Salary']])

# ==========================
# 5. Feature Extraction (PCA)
# ==========================
pca = PCA(n_components=2)
df[['PCA1', 'PCA2']] = pca.fit_transform(df[['Age', 'Salary']])

print("\nFeature Engineered Data:\n", df)
```

# 2Ô∏è‚É£ Support Vector Machine (SVM)

## ü§ñ What Is SVM?
Support Vector Machine (SVM) is a **supervised machine learning algorithm** used for **classification** and, in some cases, **regression**.  
It works by identifying an optimal **hyperplane** that maximally separates data into classes.  
Especially effective for **binary classification**, SVM employs techniques (like one-vs-rest) for handling **multi-class problems**.

---

## üîπ Why Use SVM?
- üìè **Maximizes margin** between classes‚Äîhelps generalize better on unseen data.  
- üìä **Effective in high-dimensional spaces**, even when features outnumber training samples.  
- üîÑ **Kernel trick** enables handling of *non-linear separability* by mapping data into higher-dimensional space.  
- üíæ **Memory efficient** ‚Äî operates using a subset of training points known as **support vectors**.  
- üõ°Ô∏è **Robust to outliers** through soft margin formulation, balancing accuracy and tolerance.  

---

## üåê Where Is SVM Used?
SVM shines across diverse domains:

- üìù **Text Classification**: spam detection, sentiment analysis using TF-IDF.  
- üì∑ **Image Recognition**: object detection, face recognition, digit classification.  
- üß¨ **Bioinformatics**: protein classification, gene expression analysis.  
- üí∞ **Finance**: credit scoring, stock prediction, fraud detection.  
- üè• **Medical Diagnostics**: disease classification from clinical or image data.  
- üîé **Anomaly Detection**: identifying outliers or unusual patterns.  
- üåç **Remote Sensing & Earth Sciences**: land-cover mapping, environmental monitoring; effective with limited resources.  

---

## ‚è∞ When Should You Use SVM?
- ‚úÖ Binary classification tasks with clear separation between classes.  
- ‚úÖ High-dimensional data, such as text or genomics.  
- ‚úÖ Non-linear problems (use kernels for separability).  
- ‚úÖ Noisy data or presence of outliers (soft margin helps).  
- ‚úÖ Need for memory efficiency (relies only on support vectors).  

---

## ‚ö†Ô∏è Limitations to Consider
- üê¢ Slow training times for large datasets, especially with complex kernels.  
- ‚öôÔ∏è Sensitive to parameter tuning (kernel choice, regularization).  
- ü§î Less interpretable, especially with non-linear kernels.  
- üìè Requires careful feature scaling for optimal performance.  

---

## üìã Summary Table

| **Aspect**  | **Details** |
|-------------|-------------|
| üîπ **Why**     | Builds a robust hyperplane with maximum margin for better generalization |
| üåê **Where**   | Widely used in text, image, bioinformatics, finance, medical, anomaly detection, remote sensing |
| ‚è∞ **When**    | Ideal for high-dimensional, binary or non-linear classification challenges, especially with noisy data |

# 5Ô∏è‚É£üß† Biological Inspiration: Neurons, Connections, and Weights

Neural networks are inspired by the **structure and function of the human brain**. The core elements include **neurons, connections, and weights**.

---

## üîπ Neurons
- Biological neurons are the **basic units of the brain**.
- They **receive signals** via dendrites, **process** them in the cell body, and **send output** via axons.
- In artificial neural networks:
  - Neurons are called **nodes** or **perceptrons**.
  - Each neuron **receives inputs**, applies **weights**, sums them, and passes the result through an **activation function**.
  - Output is sent to the **next layer** of neurons.

---

## üîπ Connections
- Connections in the brain = **synapses** between neurons.
- Connections in neural networks = **links between nodes** in layers.
- Each connection **transmits signal strength** based on its weight.
- Proper connectivity allows the network to **propagate information** and **learn patterns**.

---

## üîπ Weights
- Weights are **numerical values** assigned to each input connection.
- They determine the **importance of each input**.
- During training, weights are **adjusted** to minimize error (learning).
- Higher weight ‚Üí stronger influence on the neuron‚Äôs output; lower weight ‚Üí weaker influence.

---

## üîπ Summary Table

| Biological Brain | Artificial Neural Network |
|------------------|-------------------------|
| Neuron           | Node / Perceptron       |
| Synapse          | Connection              |
| Signal Strength  | Weight                  |
| Axon             | Output                  |
| Dendrites        | Inputs                  |

---

## ‚úÖ Key Points
- **Neurons**: process inputs and produce outputs.  
- **Connections**: transmit information between neurons.  
- **Weights**: control influence of inputs and are **learned during training**.  

# 6Ô∏è‚É£Perceptron Model

## 1Ô∏è‚É£ Input
- The perceptron receives multiple **input features**.  
- Each input represents a specific attribute or characteristic of the data.  
- **Example**: For classifying fruits:
  - üçé Weight: 150 grams  
  - üé® Color: 0.9 (scale 0‚Äì1, where 1 is red)  

---

## 2Ô∏è‚É£ Weighted Sum
- Each input has a **weight**, indicating its importance.  
- The perceptron calculates the **weighted sum** of all inputs plus a bias term.  
- Example: Inputs combined with weights produce a single value representing overall input strength.  
- ‚öñÔ∏è Heavier weights ‚Üí more influence on the output.
  
  

---

## 3Ô∏è‚É£ Activation Function
- Introduces **non-linearity** to the perceptron.  
- Common activation functions:
  - ‚ö° **Step Function**: Outputs 1 if input ‚â• threshold, otherwise 0  
  - üîÑ **Sigmoid Function**: Outputs between 0 and 1  
  - üÜô **ReLU**: Outputs input if positive, else 0  
- Example: Step function decides whether the input belongs to a certain class or not.  

---

## 4Ô∏è‚É£ Output
- The result from the activation function is the perceptron‚Äôs **final output**.  
- Represents the perceptron‚Äôs **decision or classification**.  
- **Example**:
  - ‚úÖ 1 ‚Üí Fruit is classified as apple  
  - ‚ùå 0 ‚Üí Fruit is classified as not apple  

---

## üìù Summary Table

| Step               | Description                                      |
|--------------------|--------------------------------------------------|
| üü¢ Input           | Receives data features                           |
| ‚öñÔ∏è Weighted Sum    | Combines inputs with weights                     |
| üîÑ Activation      | Applies a function to decide activation         |
| ‚úÖ Output          | Produces final classification or decision       |


# ‚ö†Ô∏è Perceptron Limitations

## ‚ùå Cannot Solve Non-Linear Problems
- Single-layer perceptrons are **limited to linearly separable data**.  
- Example: **XOR problem** cannot be solved by a single-layer perceptron.

---

## ‚ö° Linear Decision Boundaries Only
- Works well only when data can be separated by a straight line (or hyperplane in higher dimensions).  
- Cannot capture curved or complex decision boundaries.

---

## üê¢ Sensitive to Feature Scaling
- Large differences in input scales can affect learning.  
- Feature normalization or scaling is often required.

---

## üîÑ No Hidden Layers
- Single-layer perceptrons **cannot capture complex patterns or interactions** between features.  
- Adding hidden layers (multi-layer perceptrons) overcomes this limitation.

---

## ‚öôÔ∏è Requires Parameter Tuning
- Learning rate and threshold must be carefully chosen.  
- Improper tuning can prevent convergence or slow down learning.

---

## üí° Summary Table

| Limitation               | Explanation                                         |
|---------------------------|---------------------------------------------------|
| ‚ùå Non-linear problems    | Cannot solve XOR or other non-linearly separable data |
| ‚ö° Linear boundaries only | Decision boundary is straight, cannot capture curves |
| üê¢ Sensitive to scaling   | Requires normalization to avoid biased learning   |
| üîÑ No hidden layers       | Cannot learn complex feature interactions         |
| ‚öôÔ∏è Parameter tuning       | Learning rate and threshold need careful adjustment |

# 7Ô∏è‚É£üß† Feedforward Neural Network (FNN)

## 1Ô∏è‚É£ What Is a Feedforward Neural Network?

A **Feedforward Neural Network (FNN)** is a type of artificial neural network where information flows in a **single direction**‚Äîfrom the input layer, through any hidden layers, to the output layer. There are **no loops or cycles** in the network, meaning data moves forward only.

**Applications:**
- üñºÔ∏è Image classification  
- üéôÔ∏è Speech recognition  
- üîç Pattern recognition  
- üìä Regression tasks  

**Example:**  
In a credit scoring system, banks use an FNN to analyze users' financial profiles‚Äîsuch as income, credit history, and spending habits‚Äîto determine their creditworthiness.


## 2Ô∏è‚É£ Layers in a Feedforward Neural Network

### üîπ Input Layer
- Receives the input data.  
- Each neuron represents a **feature** of the input data.

### üîπ Hidden Layers
- One or more layers between input and output.  
- Learn **complex patterns** in the data.  
- Each neuron computes a weighted sum of inputs and applies a **non-linear activation function**.

### üîπ Output Layer
- Produces the **final output** of the network.  
- Number of neurons depends on:
  - Number of classes (classification)
  - Number of outputs (regression)

‚öôÔ∏è Each connection has a **weight** that is adjusted during training to minimize prediction errors.


## 3Ô∏è‚É£ Forward Pass (Signal Propagation)

The **forward pass** is the process of passing input data through the network to produce an output:

1. **Input Reception**  
   - Data is fed into the input layer.

2. **Weighted Sum Calculation**  
   - Each neuron calculates the weighted sum of its inputs.

3. **Activation Function Application**  
   - Weighted sum is passed through an activation function (e.g., ReLU, Sigmoid, Tanh) to introduce non-linearity.

4. **Propagation Through Layers**  
   - Output of one layer becomes input to the next layer.

5. **Final Output**  
   - Output layer produces the network‚Äôs prediction or classification.

‚úÖ During training, the forward pass is followed by **backpropagation** to update weights and minimize error.

# 8Ô∏è‚É£üìò Loss Functions in Machine Learning

Loss functions, also called cost functions, quantify the difference between predicted values and actual values.  
They guide the optimization process during training. The choice of loss function depends on the **type of problem**: regression or classification.

## 1Ô∏è‚É£ Regression: Mean Squared Error (MSE)

**Definition:**  
In regression tasks, the goal is to predict continuous numerical values. **Mean Squared Error (MSE)** measures the average squared differences between predicted and actual values.

**Formula (for reference):**  
MSE = Average of squared differences between predicted and actual values


**Characteristics:**
- ‚ö†Ô∏è Sensitive to outliers (large errors have higher impact)  
- ‚úÖ Differentiable, which is useful for optimization algorithms like gradient descent  

**When to Use:**
- When large errors are undesirable and should be penalized more heavily  
- Commonly used in linear regression and other regression models  

**Example Applications:**  
- Predicting house prices  
- Forecasting stock market prices  

## 2Ô∏è‚É£ Classification: Cross-Entropy Loss

**Definition:**  
In classification tasks, the goal is to assign data into discrete categories. **Cross-Entropy Loss** measures the difference between the true labels and predicted probabilities.

**Binary Cross-Entropy (Binary Classification):**  
- Evaluates probability of correct class (0 or 1)  
- Penalizes confident but wrong predictions  

**Categorical Cross-Entropy (Multi-class Classification):**  
- Works for multiple classes using one-hot encoded labels  
- Measures the negative log-likelihood of predicted probabilities  

**Characteristics:**
- ‚úÖ Probabilistic interpretation (works with probability outputs)  
- ‚ö†Ô∏è Penalizes confident incorrect predictions heavily  

**When to Use:**  
- Standard choice for classification problems  
- Used in logistic regression, neural networks, and deep learning models  

**Example Applications:**  
- Email spam detection  
- Image classification  

---

## üßæ Summary Table

| Task Type     | Loss Function           | Key Points                                   | Example Applications                      |
|---------------|------------------------|---------------------------------------------|------------------------------------------|
| Regression    | Mean Squared Error (MSE)| Penalizes large errors, differentiable      | House price prediction, stock forecasting |
| Classification| Cross-Entropy Loss     | Works with probabilities, penalizes wrong confident predictions | Email spam detection, image classification |


# 9Ô∏è‚É£ Vector/ Matrix operations (Dot Product = Weighted sum)

# üßÆ Vector and Matrix Operations in Neural Networks

In neural networks, inputs, weights, and outputs are often represented as **vectors** or **matrices**.  
One of the most fundamental operations is the **dot product**, which is used to calculate the **weighted sum** of inputs in a neuron.

---

## üîπ Dot Product = Weighted Sum

- Each input feature is associated with a **weight**.  
- The neuron computes a **weighted sum** of inputs to determine its total input.  
- This operation can be expressed as a **dot product** between the input vector and the weight vector.

**Mathematically:**  
If **x** = [x‚ÇÅ, x‚ÇÇ, ..., x‚Çô] is the input vector and **w** = [w‚ÇÅ, w‚ÇÇ, ..., w‚Çô] is the weight vector:



- ‚ö° This is equivalent to summing each input multiplied by its corresponding weight.  
- A **bias term** is often added:  
z = x.w+b



## üîπ Matrix Representation

- When processing multiple neurons or layers at once, inputs and weights are represented as matrices.  
- The **dot product** operation generalizes to **matrix multiplication**.

**Example:**  
- Input vector: `X` of shape (1, n)  
- Weight matrix: `W` of shape (n, m)  
- Weighted sums for all m neurons: `Z = X ¬∑ W` ‚Üí shape (1, m)

- This allows efficient computation for entire layers using linear algebra.

---

## üîπ Summary

| Concept           | Explanation |
|------------------|-------------|
| Vector            | 1D array representing inputs or weights |
| Dot Product       | Sum of element-wise multiplication (weighted sum) |
| Matrix            | 2D array representing multiple neurons or layers |
| Weighted Sum      | Dot product of input and weight vectors + bias |
| Layer Computation | Matrix multiplication allows computing all neuron outputs efficiently |

# üîü  Activation Functions in Neural Networks

Activation functions introduce **non-linearity** into the network, enabling it to learn **complex patterns**.  
Below are some commonly used activation functions.

---

## 1Ô∏è‚É£ Sigmoid Activation Function

**Definition:**  
The sigmoid function maps any input value to a range between **0 and 1**, making it useful for **binary classification tasks**.

**Formula:**  



**Characteristics:**
- Output range: 0 to 1  
- S-shaped curve  
- Smooth gradient  

**Advantages:** ‚úÖ
- Interpretable output as probability  
- Differentiable, suitable for gradient-based optimization  

**Limitations:** ‚ö†Ô∏è
- Vanishing gradient problem  
- Not zero-centered  

**Use Cases:**  
- Binary classification (e.g., spam detection)

---

## 2Ô∏è‚É£ Tanh (Hyperbolic Tangent) Activation Function

**Definition:**  
The tanh function maps input to a range between **-1 and 1**, providing a zero-centered output.

**Formula:**  



**Characteristics:**
- Output range: -1 to 1  
- S-shaped curve  

**Advantages:** ‚úÖ
- Zero-centered output  
- Stronger gradients than sigmoid  

**Limitations:** ‚ö†Ô∏è
- Vanishing gradient problem  

**Use Cases:**  
- Hidden layers in neural networks

---

## 3Ô∏è‚É£ ReLU (Rectified Linear Unit) Activation Function

**Definition:**  
Outputs the input directly if positive; otherwise, outputs zero.

**Formula:**  
f(x) = max(0, x)


**Characteristics:**
- Output range: 0 to ‚àû  
- Linear for positive inputs, zero for negative  

**Advantages:** ‚úÖ
- Computationally efficient  
- Mitigates vanishing gradient problem  

**Limitations:** ‚ö†Ô∏è
- Dying ReLU problem (neurons stop learning)  

**Use Cases:**  
- Hidden layers in deep neural networks, CNNs

---

## 4Ô∏è‚É£ Softmax Activation Function

**Definition:**  
Converts a vector of raw scores (logits) into **probabilities** by exponentiating and normalizing them.

**Formula:**  


**Characteristics:**
- Output range: 0 to 1 for each class  
- Sum of outputs = 1 (probability distribution)  

**Advantages:** ‚úÖ
- Probabilistic interpretation  
- Useful for multi-class classification  

**Limitations:** ‚ö†Ô∏è
- Sensitive to outliers  

**Use Cases:**  
- Multi-class classification (e.g., image recognition)

---

## üßæ Summary Table

| Activation Function | Output Range | Use Case                        | Advantages | Limitations |
|---------------------|--------------|---------------------------------|------------|-------------|
| Sigmoid             | 0 to 1       | Binary classification           | Smooth, interpretable | Vanishing gradient, not zero-centered |
| Tanh                | -1 to 1      | Hidden layers                   | Zero-centered, stronger gradient | Vanishing gradient |
| ReLU                | 0 to ‚àû       | Hidden layers, deep networks    | Fast, mitigates vanishing gradient | Dying ReLU problem |
| Softmax             | 0 to 1       | Multi-class classification      | Probabilistic output | Sensitive to outliers |


```python
# üß† Activation Functions in Python

import numpy as np
import matplotlib.pyplot as plt

# Input values
x = np.linspace(-10, 10, 200)

# 1Ô∏è Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 2Ô∏è Tanh Function
def tanh(x):
    return np.tanh(x)

# 3Ô∏è ReLU Function
def relu(x):
    return np.maximum(0, x)

# 4Ô∏è Softmax Function
def softmax(x):
    e_x = np.exp(x - np.max(x))  # for numerical stability
    return e_x / e_x.sum(axis=0)

# Calculate activation outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Example for softmax with 3 classes
x_softmax = np.array([2.0, 1.0, 0.1])
y_softmax = softmax(x_softmax)

# Plot Sigmoid, Tanh, ReLU
plt.figure(figsize=(12, 5))

plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.plot(x, y_tanh, label='Tanh', color='green')
plt.plot(x, y_relu, label='ReLU', color='red')

plt.title('Activation Functions')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()

print("Softmax Output:", y_softmax)
```

# 1Ô∏è‚É£1Ô∏è‚É£Optimization Techniques in Deep Learning

Optimization algorithms are responsible for reducing losses and providing the most accurate results possible. The weight is initialized using some initialization strategies and is updated with each epoch according to the equation. The best results are achieved using some optimization strategies or algorithms called Optimizers.

---

## 1Ô∏è‚É£ Gradient Descent

**Definition:**  
Gradient Descent is an iterative algorithm that starts from a random point on the function and traverses down its slope in steps until it reaches the lowest point of that function.

**Cost Function:**  



**Characteristics:**
- Entire dataset is loaded at a time.
- Computationally intensive.
- May get stuck at local minima or saddle points.

**Limitations:**
- Slow convergence.
- Computationally expensive for large datasets.

---

## 2Ô∏è‚É£ Stochastic Gradient Descent (SGD)

**Definition:**  
Stochastic Gradient Descent is an extension of Gradient Descent, where it overcomes some of the disadvantages by computing the derivative of one point at a time.

**Characteristics:**
- Computes derivatives of only 1 point at a time.
- Takes more iterations to reach minimum.
- Contains some noise compared to Gradient Descent.

**Limitations:**
- Noisy updates.
- May not converge smoothly.

---

## 3Ô∏è‚É£ Mini-Batch Stochastic Gradient Descent (MB-SGD)

**Definition:**  
MB-SGD is an extension of SGD that overcomes the time-consuming complexity by taking a batch of points/subset of points from the dataset to compute the derivative.

**Characteristics:**
- Derivative of loss function is similar to GD after some iterations.
- More iterations to achieve minima compared to GD.
- Computationally expensive.

**Limitations:**
- Noisy updates.
- May not always converge to minima.

---

## 4Ô∏è‚É£ Momentum-Based Optimizer

**Definition:**  
Momentum is an adaptive optimization algorithm that uses an exponentially weighted average of gradients over previous iterations to stabilize convergence, resulting in quicker optimization.

**Cost Function:**



**Characteristics:**
- Adds a fraction (gamma) to the previous iteration values.
- Momentum term increases when gradients point in the same direction.
- Reduces when gradients fluctuate.

**Advantages:**
- Faster convergence.
- Helps escape local minima.

---

## 5Ô∏è‚É£ Nesterov Accelerated Gradient (NAG)

**Definition:**  
NAG is an improvement over momentum-based optimization. It computes the gradient with respect to the approximate future position of the parameters.

**Cost Function:**



**Advantages:**
- Provides a more accurate estimate of the future position.
- Faster convergence compared to standard momentum.

---

## 6Ô∏è‚É£ AdaGrad

**Definition:**  
AdaGrad adapts the learning rate of parameters by updating it at each iteration depending on the position it is present.

**Characteristics:**
- Divides the learning rate by the square root of the sum of all gradients squared.
- Adapts learning rates based on feature frequency.

**Limitations:**
- Accumulation of squared gradients can cause the learning rate to shrink too much.
- May not converge to the global minimum.

---

## 7Ô∏è‚É£ AdaDelta

**Definition:**  
AdaDelta is an extension of AdaGrad that seeks to reduce its monotonically decreasing learning rate.

**Characteristics:**
- Restricts the number of summation values to a limit (w).
- Uses a decaying average of all past squared gradients.

**Advantages:**
- More robust to hyperparameter settings.
- Reduces the need for manual tuning.

---

## 8Ô∏è‚É£ RMSProp

**Definition:**  
RMSProp is another adaptive learning rate method that tries to improve AdaGrad.

**Characteristics:**
- Divides the learning rate by an exponentially decaying average of squared gradients.
- Helps maintain a more constant learning rate.

**Advantages:**
- Works well in practice.
- Suitable for non-stationary objectives.

---

## 9Ô∏è‚É£ Adaptive Moment Estimation (Adam)

**Definition:**  
Adam is a combination of RMSProp and Momentum. It computes adaptive learning rates for each parameter by storing both the first and second moments of the gradients.

**Cost Function:**



**Characteristics:**
- Combines advantages of both Momentum and RMSProp.
- Stores moving averages of both the gradients and their squares.

**Advantages:**
- Efficient computation.
- Well-suited for large datasets and parameters.

---

## üîü Nesterov Accelerated Adaptive Moment Estimator (Nadam)

**Definition:**  
Nadam is an extension of Adam that incorporates Nesterov Accelerated Gradient into the update rule.

**Advantages:**
- Combines the benefits of Adam and NAG.
- Provides faster convergence and better performance.

---

# üîÑ Backpropagation in Deep Learning

Backpropagation is a fundamental algorithm used to **train neural networks** by updating weights to minimize the loss function. It relies on **computing gradients of the loss with respect to each weight** using the **chain rule of calculus**.

---

## 1Ô∏è‚É£ What is Backpropagation?

- Backpropagation is an **efficient way to compute gradients** of a loss function for all weights in a neural network.
- It allows the network to **learn by adjusting weights** in the direction that reduces the loss.
- Works in **two main steps**:
  1. **Forward pass:** Compute the output and loss.
  2. **Backward pass:** Propagate the error backward to compute gradients using the **chain rule**.

---

## 2Ô∏è‚É£ Chain Rule of Calculus in Machine Learning

- The **chain rule** allows us to compute the derivative of a composite function.

- In neural networks:
  - Each neuron's output depends on the weights of previous layers.
  - The **chain rule helps compute gradients layer by layer**, starting from the output layer back to the input layer.


---

## 3Ô∏è‚É£ Backpropagation Intuition

1. Compute the **loss** (difference between predicted and actual values).
2. For each weight \(w\), compute **gradient of loss w.r.t weight**:


## 4Ô∏è‚É£ Why Backpropagation is Important

- ‚úÖ Efficient computation of gradients for **deep networks**.  
- ‚úÖ Allows networks to **learn complex patterns**.  
- ‚úÖ Works for any differentiable **activation function** and **loss function**.  

---

## 5Ô∏è‚É£ High-Level Summary

| Step                | Description |
|--------------------|-------------|
| Forward Pass        | Compute predictions and loss. |
| Compute Gradients   | Use **chain rule** to find how loss changes with weights. |
| Backward Pass       | Propagate errors backward layer by layer. |
| Weight Update       | Adjust weights using gradient descent or other optimizers. |

```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Input data (XOR problem)
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

# Output labels
y = np.array([[0],
              [1],
              [1],
              [0]])

# Seed for reproducibility
np.random.seed(42)

# Initialize weights randomly
input_layer_neurons = 2
hidden_layer_neurons = 2
output_neurons = 1

# Weights and biases
W1 = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))
b1 = np.random.uniform(size=(1, hidden_layer_neurons))
W2 = np.random.uniform(size=(hidden_layer_neurons, output_neurons))
b2 = np.random.uniform(size=(1, output_neurons))

# Learning rate
lr = 0.1

# Training loop
for epoch in range(10000):
    # --- Forward pass ---
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)
    
    final_input = np.dot(hidden_output, W2) + b2
    final_output = sigmoid(final_input)
    
    # --- Compute error ---
    error = y - final_output
    
    # --- Backward pass ---
    d_output = error * sigmoid_derivative(final_output)
    d_hidden = d_output.dot(W2.T) * sigmoid_derivative(hidden_output)
    
    # --- Update weights and biases ---
    W2 += hidden_output.T.dot(d_output) * lr
    b2 += np.sum(d_output, axis=0, keepdims=True) * lr
    W1 += X.T.dot(d_hidden) * lr
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * lr

# Final predictions
print("Final Output after Training:")
print(np.round(final_output, 2))
```

# 1.Implement a single-layer perceptron with NumPy

```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, lr=0.1, epochs=10):
        self.weights = np.zeros(input_size)
        self.bias = 0
        self.lr = lr
        self.epochs = epochs

    def activation(self, x):
        return np.where(x > 0, 1, 0)  # Step function

    def predict(self, x):
        linear_output = np.dot(x, self.weights) + self.bias
        return self.activation(linear_output)

    def fit(self, X, y):
        for _ in range(self.epochs):
            for xi, target in zip(X, y):
                y_pred = self.predict(xi)
                update = self.lr * (target - y_pred)
                self.weights += update * xi
                self.bias += update


# Training data for AND gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # AND truth table

# Train the perceptron
p = Perceptron(input_size=2, lr=0.1, epochs=10)
p.fit(X, y)

# Test predictions
for xi in X:
    print(f"Input: {xi}, Predicted: {p.predict(xi)}")
```

# Single-Layer Perceptron with NumPy

In this notebook, we will implement a **single-layer perceptron** from scratch using NumPy.  
We will train it on three logic gates: **AND**, **OR**, and **XOR**.  

‚ö†Ô∏è Note:  
- Perceptron can only solve **linearly separable problems** (like AND, OR).  
- It **cannot solve XOR**, since XOR is not linearly separable.

```python
# Import required libraries
import numpy as np
```

## Perceptron Class Implementation

A perceptron works as follows:

1. Takes input features **x**.  
2. Computes weighted sum
3. Applies an **activation function** (step function in this case).  
4. Updates weights using the **Perceptron Learning Rule**


```python
class Perceptron:
    def __init__(self, input_size, lr=0.1, epochs=20):
        self.weights = np.zeros(input_size)  # initialize weights to 0
        self.bias = 0                         # initialize bias to 0
        self.lr = lr                          # learning rate
        self.epochs = epochs                  # number of training iterations

    def activation(self, x):
        """Step activation function"""
        return np.where(x > 0, 1, 0)

    def predict(self, x):
        """Predict output for given input"""
        linear_output = np.dot(x, self.weights) + self.bias
        return self.activation(linear_output)

    def fit(self, X, y):
        """Train the perceptron on dataset X, y"""
        for _ in range(self.epochs):
            for xi, target in zip(X, y):
                y_pred = self.predict(xi)
                update = self.lr * (target - y_pred)  # error * learning rate
                self.weights += update * xi           # update weights
                self.bias += update                   # update bias
```

## Training Data

We will use the truth tables for **AND**, **OR**, and **XOR** gates.


```python
# Input data (same for all gates)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Output labels
y_and = np.array([0, 0, 0, 1])  # AND gate
y_or  = np.array([0, 1, 1, 1])  # OR gate
y_xor = np.array([0, 1, 1, 0])  # XOR gate
```

## Training and Testing the Perceptron

```python
# Train and test for AND
print("Training Perceptron for AND gate")
p_and = Perceptron(input_size=2, lr=0.1, epochs=10)
p_and.fit(X, y_and)

for xi in X:
    print(f"Input: {xi}, Predicted: {p_and.predict(xi)}")
print()
```

```python
# Train and test for OR
print("Training Perceptron for OR gate")
p_or = Perceptron(input_size=2, lr=0.1, epochs=10)
p_or.fit(X, y_or)

for xi in X:
    print(f"Input: {xi}, Predicted: {p_or.predict(xi)}")
print()
```

```python
# Train and test for XOR
print("Training Perceptron for XOR gate")
p_xor = Perceptron(input_size=2, lr=0.1, epochs=10)
p_xor.fit(X, y_xor)

for xi in X:
    print(f"Input: {xi}, Predicted: {p_xor.predict(xi)}")
```

## Results and Observations

- ‚úÖ The perceptron **learns AND correctly**.  
- ‚úÖ The perceptron **learns OR correctly**.  
- ‚ùå The perceptron **fails on XOR** because XOR is **not linearly separable**.

# 3.Compute loss + gradient update manually for a toy dataset.

```python
import numpy as np
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

The sigmoid activation function maps any real number into the range (0, 1).


```python
X = np.array([[0], [1], [2], [3]])
y = np.array([0, 0, 1, 1])
```

X is the input feature with 4 samples: [0],[1],[2],[3]

y is the target output: [0,0,1,1]

This means:

For input 0 ‚Üí expected output = 0

For input 1 ‚Üí expected output = 0

For input 2 ‚Üí expected output = 1

For input 3 ‚Üí expected output = 1

This mimics a step-like function (threshold behavior).

```python
w = 0.0
b = 0.0
lr = 0.1
m = len(y)
```

w: weight (starts at 0)

b: bias (starts at 0)

lr: learning rate = 0.1 (controls update step size)

m: number of samples (4 in this case)

```python
z = np.dot(X, w) + b
y_pred = sigmoid(z)
```


Since w=0 and b=0 initially:

z=0 for all inputs
y=œÉ(0)=0.5

So the model predicts 0.5 for all inputs at the start.

```python
loss = -(1/m) * np.sum(y * np.log(y_pred + 1e-9) + (1-y) * np.log(1 - y_pred + 1e-9))
print("Initial Loss:", loss)
```


It penalizes wrong predictions heavily.

The 1e-9 prevents division by zero when taking the log.

Since y=0.5 everywhere, the loss is relatively high.

```python
dw = (1/m) * np.dot(X.T, (y_pred - y))
db = (1/m) * np.sum(y_pred - y)

print("Gradient dw:", dw)
print("Gradient db:", db)
```



dw: tells how much the weight needs to change.

db: tells how much the bias needs to change.

If prediction > true label ‚Üí gradient positive ‚Üí weight decreases.
If prediction < true label ‚Üí gradient negative ‚Üí weight increases.

```python
w -= lr * dw
b -= lr * db

print("Updated weight:", w)
print("Updated bias:", b)
```


Learning rate (Œ∑) = 0.1 controls the step size.

After the update, ùë§ and b move in the direction that reduces loss.

```python
z = np.dot(X, w) + b
y_pred = sigmoid(z)
loss = -(1/m) * np.sum(y * np.log(y_pred + 1e-9) + (1-y) * np.log(1 - y_pred + 1e-9))

print("Predictions after update:", y_pred)
print("Loss after update:", loss)
```
